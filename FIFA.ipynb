## Cleaning Data

Removing 'cm' from Height and 'kg' from Weight

#### Height

* In the process of converting the values to a numeric type, it was found that several of them were expresed in the american (imperial) system.
* Using the DataFrame.replace() with a dictionary with the values converted to the metric system, they were adjusted to the former

```python
heights ={'5\'0"':'152cm',
          '5\'1"': '155cm',
          '5\'2"':'157cm',
          '5\'3"': '160cm',
          '5\'4"': '163cm',
          '5\'5"': '165cm',
          '5\'6"': '168cm',
          '5\'7"': '170cm',
          '5\'8"': '173cm',
          '5\'9"': '175cm',
          '5\'10"': '178cm',
          '5\'11"': '180cm',
          '6\'0"': '183cm',
          '6\'1"': '185cm',
          '6\'2"': '188cm',
          '6\'3"': '190cm',
          '6\'4"': '193cm',
          '6\'5"': '195cm',
          '6\'6"': '198cm',
          '6\'7"': '200m',
          '6\'8"': '203cm',
          '6\'9"': '206cm',
          '6\'10"': '208m',
          '6\'11"': '211cm'}

df.replace({'Height': heights}, inplace=True)
df['Height'] =df['Height'].str.rstrip('cm').astype(int)
df['Height'].sample(10)
```

#### Weight

```python
df[['LongName', 'Weight']].loc[df['Weight'].str.contains('lbs')]
```

```python
weights_lbs = df['Weight'].loc[df['Weight'].str.contains('lbs')].to_list()
weights_kg = [str(round((int(weight.rstrip('lbs'))/2.206)))+"kg" for weight in weights_lbs]

weights_dict = dict(zip(weights_lbs,weights_kg))

df.replace({'Weight': weights_dict}, inplace=True)

df['Weight'] = df['Weight'].str.rstrip('kg').astype(int)
df['Weight'].sample(10)
```

```python
#Converts the 'Joined' column to datetime
df['Joined'] = pd.to_datetime(df['Joined'])
df['Joined'].info()
```

```python
df['Loan Date End'].isna().value_counts()
```

```python
#Assings pd.NaT to NaN values
df['Loan Date End'] = df['Loan Date End'].fillna(pd.NaT)

# Transforms to datetime all values in the 'Loan Date End' column
df['Loan Date End'] = pd.to_datetime(df['Loan Date End'])
df['Loan Date End'].sample(10)
```

```python
#Checks the results of the dtype convertion for 'On Loan' players
df[['Contract', 'Joined','Loan Date End']].loc[df['Contract'].str.contains('On Loan')].sample(10)
```

```python
# Indexes of the player whose contract is "On Loan"
indexes = df.loc[df['Contract'].str.contains('On Loan')].index

# Creates a Series with the len players_df, assigns all values as False
on_loan = pd.Series([False for i in range(18979)])

# Assigns True to the indexes of players "On Loan"
on_loan.iloc[indexes] = True

# Joins the on_loan Series to the players_df Dataframe
df['On Loan'] = on_loan.values

# Checks with a sample from the whole dataframe
df[['Contract', 'On Loan']].sample(15)
```

```python
# Adjusting the 'Contract' column for the begining and ending years of the 'On Loan" cases

for indx in indexes:
    beg_year = df.iloc[indx]['Joined'].strftime('%Y')
    end_year = df.iloc[indx]['Loan Date End'].strftime('%Y')
    contract_str = ' ~ '.join((beg_year, end_year))
    df.loc[indx, 'Contract'] = contract_str
df.iloc[indexes]['Contract'].sample(15)
```

#### Value

```python
df['Value'].info()
```

```python
# Change the name of the column to "value_euros_millons"
df = df.rename(columns={'Value':'value_millons_of_euros'})
df['value_millons_of_euros'].info()
```

```python
#Remove the '€' symbol
df['value_millons_of_euros'] = df['value_millons_of_euros'].str.lstrip('€')
df['value_millons_of_euros'].sample(5)
```

```python
# Rename column
thousands = df['value_millons_of_euros'].loc[df['value_millons_of_euros'].str.contains('K')].to_list()

# Expose ammount in millions of euros
millions = [str((float(thousand.rstrip('K'))/1000))+"M" for thousand in thousands]
millions_dict = dict(zip(thousands, millions))
df.replace({'value_millons_of_euros': millions_dict}, inplace=True)
df['value_millons_of_euros'] = df['value_millons_of_euros'].str.rstrip('M').astype(float)
df['value_millons_of_euros'].sample(10)
```

```python
df = df.rename(columns={'Wage': 'wage_thousands_€'})
df['wage_thousands_€'] = df['wage_thousands_€'].str.lstrip('€')

thousands_k = df['wage_thousands_€'].loc[df['wage_thousands_€'].str.contains('K')].to_list()
number_thousands = [str(int(number.rstrip('K'))*1000) + 'K' for number in thousands_k]
thousands_dict = dict(zip(thousands_k, number_thousands))

df.replace({'wage_thousands_€': thousands_dict}, inplace=True)
df['wage_thousands_€'] = df['wage_thousands_€'].str.rstrip('K').astype(int)
df['wage_thousands_€'].sample(10)
```
